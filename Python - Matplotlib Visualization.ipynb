#Scatter Plot
* Scatter plot helps in visualizing 2 numeric variables. 
* It helps in identifying the relationship of the data with each variable i.e correlation or trend patterns. 
* It also helps in detecting outliers in the plot.

It is **used in **Machine learning concepts like **regression**, where x and y are continuous variables. 
It is also used in **clustering scatters or outlier detection**.

```python

insurance_df.head()
```

###Observation
The scatterplot above suggests that body mass index (BMI) and insurance charges are **positively correlated**, where **customers with higher BMI typically also tend to pay more in insurance costs**. 

(This pattern makes sense, **since high BMI** is typically associated with **higher risk of chronic disease**.)

Bar Plot shows the distribution of data over several groups.

It is used when to compare between several groups

plt.bar() takes the

### Simple Bar plot

```python
CourseTaught = ["M.Sc CS", "M.Sc. DS", "M.SC CA", "IMCA"]
Students = [63,34,70,30]
x = np.arange(len(CourseTaught))
plt.bar(CourseTaught,Students)
plt.ylabel('Student Strength')
plt.title('PG programs offered by Fergusson College')
plt.show()
```

### Horizontal Bar Plot

```python
#Horizontal
plt.barh(CourseTaught,Students)
plt.xlabel('Student Strength')
plt.title('PG programs offered by Fergusson College')
plt.show()
```

### Stacked Bar plot

```python
CourseTaught = ["M.Sc CS", "M.Sc. DS", "M.SC CA", "IMCA"]
Girls = [25, 12, 24,11]
Boys = [40,22, 40, 20]
plt.bar(CourseTaught,Boys, color = "Yellow")
plt.bar(CourseTaught,Girls, color="skyblue")
plt.title('Gender distribution PG programs offered by Fergusson College')
plt.show()
```

### Grouped Bar plot

```python
CourseTaught = ["M.Sc CS", "M.Sc. DS", "M.SC CA", "IMCA"]
Girls = [25, 12, 24,11]
Boys = [40,22, 40, 20]
x = np.arange(len(CourseTaught))
width = 0.35

## Initialize a Figure and an Axes
fig, ax = plt.subplots()
bar1 = ax.bar(x - width/2, Boys, width, label = "Boys", color="Yellow")
bar2 = ax.bar(x + width/2, Girls, width, label = "Girls", color="skyblue")
ax.set_xticks(x)
ax.set_xticklabels(CourseTaught)
ax.legend()
plt.show()
```

#Stack Plot
* Stack plot visualizes data in stacks and shows the distribution of data over time.
* It can be used for checking multiple variable area plots in a single plot.
Eg: It is useful in understanding the change of distribution in multiple variables over an interval

###Stack plot requires **stacking**, it is done in using **np.vstack()**

```python
np_data= pd.read_csv('sample_data/national_parks.csv')
np_data.head()
```

```python
x = np_data['Year']
x
```

```python
y = np.vstack([np_data['Badlands'], 
               np_data['GrandCanyon'], 
               np_data['BryceCanyon']])
y
```

```python
# Labels for each stack
labels = ['Badlands', 
          'GrandCanyon', 
          'BryceCanyon']

# Colors for each stack
colors = ['sandybrown', 
          'tomato', 
          'skyblue']

#  Similar to pandas df.plot.area()
plt.stackplot(x, y, 
              labels=labels,
              colors=colors, 
              edgecolor='black')

# Plots legend to the upperleft of Figure
plt.legend(loc=2)
plt.show()
```

##Percentage Stacked Area Chart
Stack plots based on percentage in distribution.

```python
plt.figure(figsize=(10,7))
print(np_data.sum(axis=1))
data_perc = np_data.divide(np_data.sum(axis=1), axis=0)
print(data_perc)

plt.stackplot(x,
              data_perc["Badlands"],data_perc["GrandCanyon"],data_perc["BryceCanyon"],
              edgecolor='black',
              colors=colors,
              labels=labels)

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()
```

## Boxplot using Matplotlib

```python
import matplotlib
exam_data = pd.read_csv('sample_data/exams.csv')
exam_data.head()
```

```python
#Extracting only the exam scores
exam_scores = exam_data[['math score', 'reading score', 'writing score']]
print(exam_scores.describe())
exam_scores.head()
```

```python
# For plotting in boxplot, we convert it to an array
exam_scores_array = np.array(exam_scores)
exam_scores_array
```

```python
colors = ['blue', 'grey', 'lawngreen']

bp = plt.boxplot(exam_scores_array, 
                 patch_artist=True,
                 notch=True)  #notch displays confidence interval
for i in range(len(bp['boxes'])):    
    bp['boxes'][i].set(facecolor=colors[i])    
    bp['caps'][2*i + 1].set(color=colors[i])
    
plt.xticks([1, 2, 3], ['Math', 'Reading', 'Writing'])

plt.show()
```

# Dendrogram

Dendrogram can be made with 2 types of dataset. 
*   A numeric matrix allowing to compute distance between individuals
*   hierarchical dataset where the relationship between entities is provided directly.

**info** - Ward linkage Instead of measuring the distance directly, it analyzes the variance of clusters. Ward’s is said to be the most suitable method for quantitative variables.

color_threshold  dendrogram assigns a unique color to each group of nodes in the dendrogram whose linkage is less than T

https://www.mathworks.com/help/stats/dendrogram.html


```python
mtcars_df = mtcars_df.set_index('model')
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
# Data set - mtcars
#print(mtcars_df.head())

#mtcars_df = mtcars_df.set_index('model')
plt.figure(figsize=(10,10))

# Calculate the distance between each sample
Z = linkage(mtcars_df, 'ward')

# Plot with Custom leaves, # Control number of clusters in the plot + add horizontal line. 
dendrogram(Z,  orientation='left',  leaf_font_size=8, labels=mtcars_df.index, color_threshold=15)  # color_threshold  dendrogram assigns a unique color to each group of nodes in the dendrogram whose linkage is less than T.
plt.axvline(x=240, c='grey', lw=1, linestyle='dashed')

# Show the graph
plt.show()
```

#Word Cloud
It is a cloud filled with lots of words in different sizes, which represent the frequency or the importance of each word. This is also called Tag Cloud.

##Example 1

```python
from wordcloud import WordCloud, STOPWORDS


# Create a list of word
text="Prime Minister Narendra on Friday addressed the nation highlighting the country's milestone of achieving 100 crore vaccinations. PM Modi said the world has now realised India's might after this milestone. He also said this achievement has shown power of ‘Made in India’, and reflects ‘Naya Bharat’."
text= text + "The Shiv Sena in an editorial in its mouthpiece 'Saamana' today said the government should abolish Article 370 of the Constitution, which grants special autonomous status to Jammu and Kashmir, so that people from across the nation can 'visit Kashmir and embrace its people'- alluding to Prime Minister Narendra Modi's Inde"
stopwords = set(STOPWORDS)
stopwords.update(["said", "Narendra", "Prime", "Minister", "Modi"])
# Create the wordcloud object
wordcloud = WordCloud(stopwords = stopwords, width=480, height=480, margin=0, background_color="white").generate(text)
# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear') #interpolation='bilinear' is used to make the image appear smoother.
plt.axis("off")
plt.show()
```
