## 1. Remove only punctuation that separates sentences.

```python
new_sen = ''.join(re.sub('(?<=[a-zA-Z])[\.\?\!\,](?=\s|$)','',text))
print(new_sen)
```

## 2. Extract name and email addresses 

```python
email = ', '.join(re.findall('\S+@+.\S+', text))
print(email)
name = re.findall('\([\w\s]*\)', text) 
print(name)
# '\(' for opening brakets 
# '\)' for closing brakets
# '[\w\s]*' for all character( Words ) in brakets
```

## 3. Split by delimiter \n ,\t and whitespace then lowercase all tokens.

```python
elements = re.split(r'[\n\t\s]+', text)
token_elements = [i.lower() for i in elements]
token_elements
```

## 4. Remove tokens that don't contain any alphanumeric characters: meaningless token such as [...] will be removed

```python
punch = [re.sub(r'[<>]+', '', i) for i in elements]
```
