## Methods using regular expressions

###Extract the first name from each by asking for a contiguous group of characters at the beginning of each element


```
names.str.extract('([A-Za-z]+)', expand=False)
```

###finding all names that start and end with a consonant, making use of the start-of-string (^) and end-of-string ($)

```
names.str.findall(r'[^AEIOU].*[^aeiou]$')
```

```
names1 = pd.Series(['Tannisha Kundur','Shubham Swaraj','Abhay Mehta', 'Tejas Tambe'])
names1.str.findall(r'[AEIOU].*[aeiou]$')
```
