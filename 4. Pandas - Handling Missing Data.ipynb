## Dropping null values

```
data.dropna()
```

## Dropping NA from DataFrame

```
df = pd.DataFrame([[1,      np.nan, 2],
                   [2,      3,      5],
                   [np.nan, 4,      6]])
df
```

```
df.dropna()
```

```
df.dropna(axis='columns')
```

```
#Add additional columns with all null values
df[3] = np.nan
df
```

```
#Dropping column containing all NA's
df.dropna(axis='columns', how='all')
```

```
df.dropna(axis='columns', how='any')
```

## Dropping nulls using Thresh
Thresh specifies min of null values for the row/col to be kept

```
df
df.dropna(axis='rows', thresh=3)
```

## Filling Null Values

```
data = pd.Series([1, np.nan, 2, None, 3], index=list('abcde'))
data
```

```
#fill NA entries with a single value, such as zero
data.fillna(0)
```

```
#Forward-fill to propagate the previous value
data.fillna(method='ffill')
```

```
#Back-fill to propagate the next values backward
data.fillna(method='bfill')
```
