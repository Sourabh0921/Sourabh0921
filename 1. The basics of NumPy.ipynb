# ndarray - N - dimentional array object, 
It is a fast, flexible container for large datasets.



*   Array enables to perform mathematical operations on whole blocks of data
*   It is generic multidimensional container for homogenous data.

* Every array has 
  - a **shape**, a tuple indicating the size of each dimension and a dtype
  - a **dtype**, an object describing the data type of the array




# Creating Arrays

```python
# integer array:
np.array([1, 4, 2, 5, 3])
```

```python
data = [[1,2,3],[4,5,6]]
arr = np.array(data)
arr
```

```python
#dimensions
arr.ndim
```

```python
#shape
arr.shape
```

```python
#datatype
arr.dtype
```

## Array Creation Functions

image.png

```python
# Create a length-10 integer array filled with zeros
np.zeros(10, dtype=int)
```

```python
# Create a 3x5 floating-point array filled with ones
np.ones((3, 5), dtype=float)
```

```python
# Create an array filled with a linear sequence
# Starting at 0, ending at 20, stepping by 2
# (this is similar to the built-in range() function)
np.arange(0, 20, 2)
```

```python
# Create a 3x3 identity matrix
np.eye(3)
```

```python
np.full((2,3),fill_value=4)
```
