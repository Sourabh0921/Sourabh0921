```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
penguins_data="https://raw.githubusercontent.com/datavizpyr/data/master/palmer_penguin_species.tsv"
penguins_df = pd.read_csv(penguins_data, sep="\t")
penguins_df.head()
```

##Scatterplot

The set_context () allows us to override default parameters. This affects things like the size of the labels, lines, and other elements of the plot, but not the overall style.

```python
sns.set_context("talk", font_scale=1.1)
plt.figure(figsize=(8,6))
sns.scatterplot(x="culmen_length_mm", 
                y="culmen_depth_mm", 
                data=penguins_df)
plt.xlabel("Culmen Length (mm)")
plt.ylabel("Culmen Depth (mm)")
plt.savefig("Seaborn_scatterplot.png",
                    format='png',dpi=150)
```

##Bubble plot
Its special type of Scatter plot with size specified

```python
sns.set_context("talk", font_scale=1.1)
plt.figure(figsize=(10,6))
sns.scatterplot(x="culmen_length_mm", 
                y="culmen_depth_mm",
                size="body_mass_g", 
                alpha=0.5,
                hue="sex",           
                data=penguins_df,
                legend="brief")
# Put the legend out of the figure
plt.legend(bbox_to_anchor=(1.01, 1),borderaxespad=0)
# Put the legend out of the figure
#plt.legend(bbox_to_anchor=(1.01, 0.54),  borderaxespad=0.)
plt.xlabel("Culmen Length (mm)")
plt.ylabel("Culmen Depth (mm)")
plt.title("Bubble plot in Seaborn")
plt.tight_layout()
```
