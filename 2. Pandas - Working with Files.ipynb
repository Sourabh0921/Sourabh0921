#Working with Data

### Reading certain columns for a range of rows

```
print(data.loc[:,['title','rating']])
```

### Reading certain rows and certain columns 

```
print(data.loc[[0,1],['title','rating']])
```

```
df = pd.DataFrame([[1, 2], [4, 5], [7, 8]],
     index=['cobra', 'viper', 'sidewinder'],
     columns=['max_speed', 'shield'])
df
```

### Single Label

```
df.loc['viper']
```

### List of labels using [[]] , returns Data frame


```
df.loc[['viper', 'sidewinder']]
```

## Single label for row and column

```
df.loc['cobra', 'shield']
```

## Slice with labels for row and single label for column

```
df.loc['cobra':'viper', 'max_speed']
```

## Boolean list with the same length as the row axis

```
df.loc[[False, False, True]]
```

## Conditinal retrievals

```
a = df.loc[df['shield'] > 4]
a
```
