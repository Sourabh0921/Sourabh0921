## Ufuncs: Index Preservation

```
import pandas as pd
import numpy as np
```

```
#Ufuncs preserve indexes after performing operations on padas objects
rng = np.random.RandomState(42)
ser = pd.Series(rng.randint(0, 10, 4))
print(ser)
```

```
np.exp(ser)
```

```
df = pd.DataFrame(rng.randint(0, 10, (3, 4)),
                  columns=['A', 'B', 'C', 'D'])
print(df)
```

```
np.sin(df * np.pi / 4)
```

## Index alignment in Series

```
area = pd.Series({'Alaska': 1723337, 'Texas': 695662,
                  'California': 423967}, name='area')
population = pd.Series({'California': 38332521, 'Texas': 26448193,
                        'New York': 19651127}, name='population')
print(area)
print(population)
```

```
print(population)
```

```
a = population / area
a

#NaN is introduced for missing index values
```

## Indices handling missing values 

```
A = pd.Series([2, 4, 6], index=[0, 1, 2])
B = pd.Series([1, 3, 5], index=[1, 2, 3])
print(A)
print(B)
```

```
A + B
```

```
# Incase of missing value, fill_value can be used to perform default operation
A.add(B, fill_value=0)
```

## Index alignment in DataFrame

```
A = pd.DataFrame(rng.randint(0, 20, (2, 2)), columns=list('AB'))
B = pd.DataFrame(rng.randint(0, 10, (3, 3)), columns=list('BAC'))

print(B)
```

```
print(A)
```

```
A+B
# Irrespective of order of columns, they are matched and + operation is performed
```

```
A.add(B, fill_value=0)
```

```
#rather than default fill = 0, populate mean of A and fill
fill = A.stack().mean()
A.add(B, fill_value=fill)
```

```
fill
```

## Ufuncs: Operations Between DataFrame and Series

```
#the index and column alignment is maintained while performingn operation between 2D and 1D array
A = rng.randint(10, size=(3, 4))
A
```

```
A - A[0]
```

```
B = [1,7,5,1]
A - B
```

```
df = pd.DataFrame(A, columns=list('QRST'))
df
```

```
df - df.iloc[0]
```

```
df.iloc[0]  #0th row
print(df)
```

```
# Specfying column level operations
# use Object methods
df.subtract(df['R'], axis=0)
```

```
halfrow = df.iloc[0, ::2]
halfrow
```

```
df - halfrow
```
